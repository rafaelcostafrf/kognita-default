```python
import pandas as pd
import seaborn as sns
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np

# Importando a database
df = pd.read_csv('_data/dataset_2021-5-26-10-14.csv', sep='\t', encoding = 'utf-8')

display(df.head())
```

```python


def plot_hist(df_0, df_1, col, title, bins):
    '''plot_hist [summary]

    Arguments:
        df_0 {dataframe} -- dataframe com default zero
        df_1 {dataframe} -- dataframe com default um
        col {str} -- nome da coluna
        title {str} -- título da figura
        bins {list} -- cestos de separação dos dados
    '''    
    plt.hist([df_0[col], df_1[col]], 
            color=['r','b'], 
            alpha=0.9, 
            bins=bins,
            label=['Default 0', 'Default 1'])


    plt.xlabel(col, fontsize=16)  
    plt.ylabel("Quantidade")
    plt.suptitle(title)
    plt.legend()
    plt.show()

# Configurando o tamanho da figura do seaborn
sns.set(rc={'figure.figsize':(12, 12)})

# Número de entradas na database
print(f'Número de entradas: {len(df.index):d}')

# Usando describe para analisar os dados das colunas
described_df = df.describe()
display(described_df)

# A média de default
print(f'Probabilidade média de default por pedido: {described_df["default"]["mean"]:.3f}')

# Histograma de valores dos pedidos
print('Histograma de valores de pedidos')
ax_d = sns.displot(df['valor_total_pedido'], bins=np.arange(0, 15000, 1000), kde=False)
plt.show()

# Histograma de valores de pedidos separados por default 1
df_default = df.where(df['default']==1)

# Histograma de valores de pedidos separados por default 0
df_n_default = df.where(df['default']==0)

plot_hist(df_0=df_n_default, 
            df_1=df_default, 
            col='valor_total_pedido', 
            title='Distribuição do valor total do pedido e default', 
            bins=np.arange(0, 15000, 1000))

plot_hist(df_0=df_n_default, 
            df_1=df_default, 
            col='default_3months', 
            title='Distribuição de População entre default nos últimos 3 meses e default no pedido atual', 
            bins=np.arange(0, 4, 1))

# Distribuição de Valor de pedido, excluindo outliers
ax = sns.boxplot(y=df['valor_total_pedido'], x=df['default'], showfliers=False)
```

```python
# Análise de valor de default
grouped_default = df.groupby(['default']).sum()
rec = grouped_default["valor_total_pedido"][0]
default = grouped_default["valor_total_pedido"][1]
print(f'Valor Recebido: {rec:.2f}')
print(f'Valor Default: {default:.2f}')
print(f'Porcentagem de valor perdido: {default/(rec+default):.2%}')

# Análise de default por tipo de sociedade
d_tipo = pd.DataFrame({'default 0': df_n_default['tipo_sociedade'].value_counts(), 
                        'valor recebido': df_n_default.groupby('tipo_sociedade')['valor_total_pedido'].sum(), 
                        'default 1': df_default['tipo_sociedade'].value_counts(),
                        'valor default': df_default.groupby('tipo_sociedade')['valor_total_pedido'].sum(),})
d_tipo = d_tipo.sort_values(by='valor recebido', ascending=False)
d_tipo[['default 0', 'default 1']].plot(kind='bar', subplots=False, )
plt.show()

# Função para cálculo da probabilidade de defalt
def porcentagem_default(x, y):
    if np.isnan(x) or np.isnan(y):
        return 0
    else:
        return 100 * x / (x+y)

def porcentagem_prejuizo(x, y):
    if np.isnan(x) or np.isnan(y):
        return 0
    else:
        return 100 * x / (y)


valor_total_perdido = df['valor_total_pedido'].sum()
d_tipo['porcentagem default'] = d_tipo.apply(lambda x: porcentagem_default(x['default 1'], x['default 0']), axis=1)
d_tipo['porcentagem prejuizo'] = d_tipo.apply(lambda x: porcentagem_prejuizo(x['valor default'], d_tipo['valor default'].sum()), axis=1)
display(d_tipo.sort_values(by='porcentagem prejuizo', ascending=False))
```

Nesta análise inicial estamos observando uma tabela com 117.000 entradas. A probabilidade média de default de um pedido é de 16.7% e mais de 75% dos pedidos foram realizados por clientes que nao têm defaults nem protestos, obsevando as colunas 'defaults_3_months' e 'quand_protestos' na tabela de descrição da database.

Podemos também observar no gráfico acima que o valor de pedido médio do default é ligeiramente maior em comparação ao valor de pedidos que não geram default, demonstrando que pedidos com maior valor tem maior chance de serem default.


```python
# correlação entre colunas
correlacao = df.corr()

# heatmap de correlação
sns.heatmap(correlacao, 
        xticklabels=correlacao.columns,
        yticklabels=correlacao.columns,
        cmap="YlGnBu")
```

Podemos observar no gráfico de correlação acima de que as colunas centrais têm certa correlação entre elas, como por exemplo a forte correlação entre valor por vencer, com valor vencido e valor quitado. Default, que é a variável e interesse neste caso, tem baixa correlação com os dados da tabela, com exceção da coluna de defaults nos últimos 3 meses.
